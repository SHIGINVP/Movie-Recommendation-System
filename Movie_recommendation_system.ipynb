# Data Preprocessing

```python
import numpy as np
import pandas as pd
import ast
```

```python
credits = pd.read_csv('/content/data/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/data/tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
credits.info()
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
# merging both the dataframe
movies = movies.merge(credits, left_on='title', right_on='title')
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(2)
```

```python
# Convert "genres" column into a list of genre names
movies["genres"] = movies["genres"].apply(lambda x: [genre["name"] for genre in ast.literal_eval(x)])
```

```python
movies.head(2)
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x: [keyword["name"] for keyword in ast.literal_eval(x)])
```

```python
movies.head(2)
```

```python
movies.cast[0]
```

```python
movies["cast"] = movies["cast"].apply(lambda x: [cast["name"] for cast in ast.literal_eval(x)[:3]]) # Only top 3 actors
```

```python
movies.head(2)
```

```python
movies.crew[0]
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies.tags[0]
```

```python
movies = movies[['movie_id', 'title', 'overview', 'tags']]
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies.head()
```

# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
```

# Checking similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar movies
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```

```python
print(get_recommendations('The Dark Knight Rises'))
```

```python
import pickle
with open('movie_data.pkl', 'wb') as file:
    pickle.dump((movies, cosine_sim), file)
```
